```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
crime = pd.read_csv('avocado.csv')
crime.head()
```

```python
import pandas as pd 

type = {'conventional': 1,'organic': 0} 
  
crime.type = [type[item] for item in crime.type] 
print(crime)
```

```python
X = crime.drop(['type','Date','region'], axis=1)
X.head()
```

```python
y = crime["type"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
y_pred_train = logreg.predict(X_train)  
```

```python
y_pred_test = logreg.predict(X_test)                                                           # make predictions on the testing set
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy score for test data is:', accuracy_score(y_test,y_pred_test))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = pd.DataFrame(confusion_matrix(y_test, y_pred_test))

print(confusion_matrix)
```

```python
confusion_matrix.index = ['actual convo','actual organ']
confusion_matrix.columns = ['pred convo','pred organ']
print(confusion_matrix)
```

```python
preds1 = np.where(logreg.predict_proba(X_test)[:,1]> 0.75,1,0)
print('Accuracy score for test data is:', accuracy_score(y_test,preds1))
```

```python
preds2 = np.where(logreg.predict_proba(X_test)[:,1]> 0.25,1,0)
print('Accuracy score for test data is:', accuracy_score(y_test,preds2))
```
